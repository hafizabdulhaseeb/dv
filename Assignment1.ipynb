# Line Plot

```python
# Datasets source: "https://data.world/uscensusbureau/construction-price-indexes"

import pandas as pd
from matplotlib import pyplot as plt
fig = plt.figure(figsize=(18,6))
ax = fig.add_subplot()
plt.style.use('fivethirtyeight')

x = []
y = []
# The function import dataset and make line plot x and y values
def line_plot(i):
    
    data = pd.read_csv('pricesHouses.csv')
    x =data['Date']
    y1 = data['Laspeyres (Fixed)']
    y2 = data['Fisher (Deflator)']
    ax.plot(x, y1, label='Laspeyres (Fixed)')
    ax.plot(x, y2, label='Fisher (Deflator)')
    plt.legend(loc='upper right')
    plt.tight_layout()
    plt.title('Price Indexes of New Single-Family Houses Under Construction')
    ax.set_xlabel('Dates')
    ax.set_ylabel('Prices Indexes')

line_plot(1)
plt.tight_layout()
plt.grid(True)
plt.show()
```

# Bar Plot

```python
# Datasets source: "https://data.world/makeovermonday/2018w50"
import pandas as pd
from matplotlib import pyplot as plt
fig = plt.figure(figsize=(18,6))
ax = fig.add_subplot()
plt.style.use('fivethirtyeight')

x = []
y = []
# The function import dataset and make bar plot x and y values
def bar_plot(i):
    
    data = pd.read_csv('barplot.csv')
    foodType = data['Food Type']
    landUse = data['Land Use (in sqm)']
    ax.bar(x=foodType, height=landUse, width=0.8, label='Land Used')
#     ax.plot(x, y2, label='Fisher (Deflator)')
    plt.legend(loc='upper right')
    plt.tight_layout()
    plt.title('Land use by food type in 2017')
    ax.set_xlabel('Food Types')
    ax.set_ylabel('Land Use (in sqm)')

bar_plot(1)
plt.tight_layout()
plt.grid(True)
plt.show()
```

# Scatter PLot

```python
# Datasets source: "https://www.kaggle.com/datasets/mirichoi0218/insurance/code"
import pandas as pd
from matplotlib import pyplot as plt
fig = plt.figure(figsize=(18,6))
ax = fig.add_subplot()
plt.style.use('fivethirtyeight')

x = []
y = []
# The function import dataset and make scatter plot x and y values
def scatter_plot(i):
    
    data = pd.read_csv('insurance.csv')
    BMI = data['bmi']
    Charges = data['charges']
    Smokers = data['smoker']
    ax.scatter(x=BMI, y=Charges)
#     ax.plot(x, y2, label='Fisher (Deflator)')
    plt.legend(loc='upper right')
    plt.tight_layout()
    plt.title('Relationship Between BMI and Charges')
    ax.set_xlabel('bmi')
    ax.set_ylabel('charges')

scatter_plot(1)
plt.tight_layout()
plt.grid(True)
plt.show()
```
